```python
from collections import Counter
import operator


#
# Function returns the maximally occuring k-mer of length k (input) for the input sequence
#

def get_number_of_kmers(seq,k):
    kmer_dict = dict(Counter([seq[i:i+k] for i in range(len(seq)-k)]))
    
    max_key = max(kmer_dict, key=lambda k: kmer_dict[k])   #select the sequence with the most k-mer hits
    #note: if all kmers in the kmer dict have same # of hits, this will just return the first kmer = BAD
    
    output = None  

    #double-check that selected k-mer really is larger than all others (see NOTE above)
    if(sum([kmer_dict[max_key] > i for i in kmer_dict.values()]) > 0):          
        if(len(set(max_key)) > 1):  #check that the kmer has > 1 occurrence.

            #print(seq)
            #print(max_key)     
            output = max_key #set output to the "max kmer"
            
    #if the kmer is not larger than the others, there is no obvious winner...return 'None' for this length k
            
    return(output)
```

```python
# Open the final file and loop through all subsequences that occurred > 1 time...if you find the subsequence
# in the original file, move on. If not, write it to a list of "failed_sequences" [read: unrecoverable sequences].

full_file = open('../PFpeptides.chopped.hpr8.xtoa.lzw40.fasta','r').read()

failed_sequences = []  #sequences for which we were unable to recover their repeat element elsewhere in the file

count = 0
for k in results.keys():  #loop through all sequences in the results dictionary 
    
    #flag variable - are any of the maximally occurring repetetive k-mers present elsewhere in the saved file?
    found_again = False  
    
    if all(v is None for v in results[k]):  #there are no obvious winning kmers for any length k...don't continue checking
        continue
        #print('x')

    else:
        for j in results[k]:  #LOOP over all possible k-mers for the associated sequence

            if j:   #make sure values is not None
                if j in full_file:   #check if the kmers are present in the full_file
                    found_again = True

        if not found_again:  #if none of the sequence-specific double-kmers were found in the file, print the seq.
            print(k)  #any seq printed here is one that we would lose.
            print(results[k])
            count += 1
            failed_sequences.append(k)
            
#print the final percentage of all sequences for which the most abundant repeat element was not recoverable in 
#another location in the final sequences
print(count)
print(len(results))
print(count/len(results))

# .8% of all hpr sequences that were removed are unrecoverable
# None (0/159) of the lzw40 sequences that were removed are unrecoverable
```
